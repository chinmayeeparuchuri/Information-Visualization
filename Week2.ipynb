# Random Sampling

```python
import random

population = list (range(1,22))
sample_size = 21
random_sample = random.sample(population, sample_size)
print("random Sample: ", random_sample)
```

# Strafield Sampling

```python
#strafield Sampling
import numpy as np
import pandas as pd

data = pd.DataFrame({
    'Category' : ['A', 'B', 'C', 'D'] * 25,
    'value' : np.random.randn(100)
    })

strafield_sample = data.groupby('Category', group_keys= False).apply(lambda x: x.sample(2))
print("strafield Sample: \n", strafield_sample)
```

# Cluster Sampling

```python
#cluster Sampling
clusters = [list(range(1,26)), list(range(26,51)), list(range(51,76)), list(range(76,100))]
cluster_sample = random.choice(clusters)
print("Cluster Sample", cluster_sample)
```

# Data Sampling and dimensionality reduction
# Principle Composite Analysis

```python
# data Sampling and Dimensionality reduction
# Principal composite Analysis (PCA)
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
iris = load_iris()
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(iris.data)

plt.scatter(reduced_data[:,0], reduced_data[:,1], c=iris.target, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')
plt.colorbar(label="Species")
plt.show()
```

# Parallel Coordinates Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates
import requests
from io import StringIO

# Fetch the data using requests
url = 'https://raw.githubusercontent.com/plotly/datasets/master/iris-data.csv'
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    data = pd.read_csv(StringIO(response.text))
    
    # Plotting the parallel coordinates
    parallel_coordinates(data, 'class', color=['b','r','g'])
    plt.title('Parallel Coordinates Plot')
    plt.xlabel('Attributes')
    plt.ylabel('Values')
    plt.show()
```

# Time Series Visualization

```python
import plotly.express as px
import pandas as pd

# Sample Time Series Data
data = pd.DataFrame({
    'Year': [2000, 2001, 2002, 2003, 2004, 2005],
    'Value': [100, 200, 150, 300, 250, 400],
    'Country': ['USA']*6
})

# Create a time series plot using Plotly Express
fig = px.line(data, x='Year', y='Value', title='Time Series Data Visualization', labels={'Value': 'Sample Value'})
fig.show()
```
